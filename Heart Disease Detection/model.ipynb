### Separating independent and dependent varibale

```python
x = df.drop(columns='target', axis=1)
y = df['target']
```

### Splitting the data into training and testing purpose

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=20)
```

```python
print(x_train.shape , x_test.shape , y_train.shape, y_test.shape)
```

# Model Building

I will be experimenting with Three Algorithm

1. Logistic Regression
2. Support Vector Machine
3. Random Forest Classifier
4.  Decision Tree Classifier

### Logistic Regression Model

```python

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
lr_model = LogisticRegression()
lr_model.fit(x_train,y_train)
```

```python
y_pred=lr_model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score

print(confusion_matrix(y_test,y_pred))

print("-"*50)
ac_lr = accuracy_score(y_test,y_pred)
ac_lr
```

### Support Vector Machine

```python
from sklearn.svm import SVC
```

```python
svc = SVC()
svc.fit(x_train,y_train)
```

```python
y_pred_svc=svc.predict(x_test)
cm_svc= confusion_matrix(y_test,y_pred_svc)
cm_svc
```

```python
ac_svc= accuracy_score(y_test,y_pred_svc)
ac_svc
```

### Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=90,random_state=0)
rf.fit(x_train,y_train)
```

```python
y_pred_rf=rf.predict(x_test)
cm_rf=confusion_matrix(y_test,y_pred_rf)
print(cm_rf)

print("-"*50)

ac_rf = accuracy_score(y_test,y_pred_rf)
ac_rf
```

### Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier(criterion='entropy',random_state=0)
dt.fit(x_train,y_train)
```

```python
y_pred_dt = dt.predict(x_test)

print("-"*50)
cm_dt = confusion_matrix(y_test,y_pred_dt)
print(cm_dt)


print("-"*50)

ac_dt = accuracy_score(y_test,y_pred_dt)
ac_dt
```

```python
algorithms = ['Logistic Regression'," Support Vector Classification","Random Forest Classifier","Decison Tree Classifier"]
accuracyscore = [ac_lr*100,ac_svc*100,ac_rf*100,ac_dt*100]
```

```python
accuracy_table = pd.DataFrame({'Algorithms':algo,'Accuracy score (In %)':accuracyscore})
accuracy_table
```

```python
accuracy_sort =  accuracy_table.sort_values(['Accuracy score (In %)'],ascending=False)
sns.barplot(x =accuracy_sort['Algorithms'].values,y=accuracy_sort['Accuracy score (In %)'].values)
plt.xticks(rotation='vertical')
plt.title("Accuracy Score of Each Algorithm")
```
