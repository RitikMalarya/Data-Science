# Data Visualization

```python
sns.pairplot(data=data)
```

```python
total_cases = data.sum()
total_cases
```

```python
total_cases.drop(['sno','state_name','state_code'])
```

```python
# visulaize Number of Active cases in Each State
data[['state_name','active']].groupby(["state_name"]).mean().sort_values(by='active').plot.bar(color='purple')
plt.show()
```

```python
# visualize no. of death in each state
data[['state_name','death']].groupby(["state_name"]).mean().sort_values(by='death').plot.bar(color='purple')
plt.show()
```

```python
# Visualize no of active cases in each state
plt.figure(figsize=(13,8))
plt.xticks(rotation=90)
sns.barplot(x='state_name',y='active',data=data)
```

```python
# plotting Graph  no. of postive cases in Each state
plt.figure(figsize=(13, 8))
plt.xticks(rotation=90)
sns.barplot(x='state_name',y='positive',data=data) 
```

```python
# visualize no. of deaths in each state
plt.figure(figsize=(13, 8))
plt.xticks(rotation=90)
sns.barplot(x='state_name',y='death',color='blue',data=data)
```

# Pie Chart Visualization|

```python
my_data = [820096,65977346,64276184,8810656]
my_labels = 'Active','Positive','Cured','Death'
my_explode = (0,0,0,1.5)
my_colors = ['lightblue','lightsteelblue','lightpink','black']
fig1, ax1 = plt.subplots(figsize=(13, 8))
plt.pie(my_data, labels=my_labels, autopct='%1.1f%%', startangle=15, shadow = True, colors=my_colors, explode=my_explode)
plt.axis('equal')
plt.show()
# Pie chart visualization
```

```python
my_data = [809748,66055242,64363990,881504]
labels = 'New Active','New Positive','New Cured','New Death'
explode = (0,0,0,1.5)
colors = ['Yellow','Green','lightpink','black']
fig1, ax1 = plt.subplots(figsize=(13, 8))
plt.pie(my_data, labels=labels, autopct='%1.1f%%', startangle=15, shadow = True, colors=my_colors, explode=my_explode)
plt.axis('equal')
plt.show()
```
